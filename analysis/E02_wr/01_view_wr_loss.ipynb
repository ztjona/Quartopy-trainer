```python
import plotly.graph_objs as go

win_rate = evals["win_rate"]

fig = go.Figure()
for bot, rates in win_rate.items():
    fig.add_trace(go.Scatter(y=rates, mode="lines", name=bot))

fig.update_layout(
    title="Win Rate per Epoch",
    xaxis_title="Epoch",
    yaxis_title="Win Rate",
    legend_title="Bot",
)
fig.show()
```

```python
loss_epoch = [
    evals["loss_values"]["loss_values"][j] for j in evals["loss_values"]["epoch_values"]
]
```

```python
import plotly.graph_objs as go

win_rate = evals["win_rate"]
w_r_smoothed = {}
for bot, rates in win_rate.items():
    smoothed = []
    for i in range(len(rates)):
        window = rates[max(0, i - 5) : min(len(rates), i + 6)]
        smoothed.append(sum(window) / len(window))
    w_r_smoothed[bot] = smoothed

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        y=win_rate["bot_good"],
        mode="markers",
        name="bot_good",
        marker=dict(color="blue", size=3),
    )
)
fig.add_trace(
    go.Scatter(
        y=w_r_smoothed["bot_good"],
        mode="lines",
        name="good_smoothed",
        line=dict(width=3, color="blue"),
    )
)
fig.add_trace(
    go.Scatter(
        y=loss_epoch,
        mode="lines",
        name="Loss",
        # line=dict(width=3, color="blue"),
    )
)

fig.add_vline(x=27, line_width=2, line_dash="dash", line_color="red")
fig.update_layout(
    title="Win Rate per Epoch",
    xaxis_title="Epoch",
    yaxis_title="Win Rate",
    xaxis=dict(range=[0, 200]),  # Set your desired limits here
)
fig.show()
```
