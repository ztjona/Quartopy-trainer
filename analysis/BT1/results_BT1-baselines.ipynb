```python
import pandas as pd

# df = pd.DataFrame(p_i.values(), index=p_i.keys())
df_BT = pd.DataFrame.from_dict(p_i, orient="index", columns=["BT"])
# Display the DataFrame
df_BT.drop([-1, 0], inplace=True)
print(df_BT)
```

```python
# calc win rate
def calc_win_rate(baseline_vs):
    w_rate_i_vs_baseline = {}
    for i, res in baseline_vs.items():
        total_games = res["wins"] + res["losses"] + res["draws"]
        if total_games > 0:
            w_rate = (res["losses"] + res["draws"] * 0.5) / total_games
        else:
            w_rate = 0
        w_rate_i_vs_baseline[i] = w_rate

    return w_rate_i_vs_baseline
```

```python
w_rate_i_vs_good = calc_win_rate(good_vs)
w_rate_i_vs_random = calc_win_rate(random_vs)
w_rate_i_vs_michael = calc_win_rate(michael_vs)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(
    w_rate_i_vs_good, orient="index", columns=["win_rate_vs_good"]
)
df["win_rate_vs_random"] = pd.Series(w_rate_i_vs_random)
df["win_rate_vs_michael"] = pd.Series(w_rate_i_vs_michael)
df
```

```python
import numpy as np

window = 100
vs_random_smooth = np.convolve(
    df["win_rate_vs_random"], np.ones(window) / window, mode="same"
)
vs_good_smooth = np.convolve(
    df["win_rate_vs_good"], np.ones(window) / window, mode="same"
)
vs_michael_smooth = np.convolve(
    df["win_rate_vs_michael"], np.ones(window) / window, mode="same"
)
df["vs_random_smooth"] = pd.Series(vs_random_smooth)
df["vs_good_smooth"] = pd.Series(vs_good_smooth)
df["vs_michael_smooth"] = pd.Series(vs_michael_smooth)
df
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df,
    x=df.index,
    y="win_rate_vs_good",
    marker="o",
    linestyle=":",
    label="vs Good",
)
sns.lineplot(
    data=df,
    x=df.index,
    y="win_rate_vs_random",
    marker="o",
    linestyle=":",
    label="vs Random",
)
plt.legend()
plt.title("Win Rate vs Good")
plt.xlabel("Index")
plt.ylabel("Win Rate")
plt.show()
```
