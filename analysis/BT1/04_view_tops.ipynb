```python
import pickle

file = "eval_BT01__tops_results.pkl"

with open(file, "rb") as f:
    _evals = pickle.load(f)
```

```python
import pandas as pd

players = _evals.keys()
results = pd.DataFrame(0.0, columns=players, index=players)

for pi, _res_pi in _evals.items():
    for pj, _resp_pij in _res_pi.items():
        results.loc[pi, pj] += _resp_pij["wins"] + 0.5 * _resp_pij["draws"]
        results.loc[pj, pi] += _resp_pij["losses"] + 0.5 * _resp_pij["draws"]


results
```

```python
from utils import calculate_BradleyTerry

score = {i: 1.0 for i in players}
calculate_BradleyTerry(
    score, results, EPOCHS=10, normalize=True, verbose=True, diff_threshold=1e-6
)
```

```python
import pickle

file = "p_i.pkl"

with open(file, "rb") as f:
    p_i_ = pickle.load(f)
BT_old = {i: p_i_[i] for i in players}
BT_old
```

```python
correlation = pd.Series(BT_old).corr(pd.Series(score))
print("Correlation between BT_old and score:", correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(list(BT_old.values()), list(score.values()), alpha=0.7)
plt.xlabel("BT_old")
plt.ylabel("score")
plt.title("BT_old vs score")
plt.grid(True)
plt.show()
```
