```python
from collections import defaultdict
import pandas as pd

results = pd.DataFrame(
    [[0, 2, 0, 1], [3, 0, 5, 0], [0, 3, 0, 1], [4, 0, 3, 0]],
    columns=list("ABCD"),
    index=list("ABCD"),
)
results
```

```python
results["A"]["B"]
```

```python
# Access wins of i over j
w = lambda i, j: results[j][i]
w("A", "B")
```

```python
p_i = [1, 1, 1, 1]
EPOCHS = 5

from math import prod


for e in range(EPOCHS):
    for i, iL in enumerate("ABCD"):
        _num = 0
        _den = 0
        for j, jL in enumerate("ABCD"):
            if i == j:
                continue

            _num += w(iL, jL) * p_i[j] / (p_i[i] + p_i[j])
            _den += w(jL, iL) / (p_i[i] + p_i[j])
        p_i[i] = _num / _den
        _m_g = prod(p_i) ** (0.25)
        p_i = [x / _m_g for x in p_i]
    print([f"{x:.3f}" for x in p_i])
```
