```python
_results
# Compute Bradley-Terry scores as agents are added incrementally
bt_scores = []
p = [1.0, 1.0]  # Initial scores with 2 agents
for n in range(2, len(_results) + 2):  # Start from 2 agents up to N_AGENTS + 1
    # Build results matrix for first n agents as a list of lists
    partial_results = [[0.0 for _ in range(n)] for _ in range(n)]
    for i, _res_i in enumerate(_results[: n - 1]):
        for j, _res_ij in _res_i.items():
            if j >= n:
                continue
            partial_results[i + 1][j] += _res_ij["wins"]
            partial_results[j][i + 1] += _res_ij["losses"]
            partial_results[i + 1][j] += _res_ij["draws"] / 2
            partial_results[j][i + 1] += _res_ij["draws"] / 2

    # Add dummy player for regularization
    for k in range(n):
        partial_results[n - 1][k] = 1.0
        partial_results[k][n - 1] = 1.0
    partial_results[n - 1][n - 1] = 0.0

    # Bradley-Terry iterative update
    # Use the previous p, pad with 1.0 if needed
    prev_p = list(bt_scores[-1])
    if len(prev_p) < n + 1:
        prev_p += [1.0] * (n + 1 - len(prev_p))
    p = prev_p

    def w(i, j):
        return partial_results[j][i]

    for _ in range(EPOCHS):
        p_old = p.copy()
        for i in range(n):
            num = 0
            den = 0
            for j in range(n):
                if i == j:
                    continue
                num += w(i, j) * p[j] / (p[i] + p[j])
                den += w(j, i) / (p[i] + p[j])
            p[i] = num / den if den != 0 else 1.0
    bt_scores.append(p[:n])  # Exclude dummy

bt_scores  # List of BT scores as agents are added
```
